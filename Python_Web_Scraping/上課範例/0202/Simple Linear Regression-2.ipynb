```python
# 作業:簡單線性回歸(Simple Linear Regression): 由現有的資料建立一條線, 從這條線可以看出資料的相關性, 而獲得對未來發展的預測。
# 適合：預測, 資料之間的變項有相關性
```

```python
# 高中生合理的身高與體重
# 身高：147.9, 163.5, 159.8, 155.1, 163.3, 158.7, 172.0, 161.2, 153.9, 161.6
# 體重：41.7, 60.2, 47.0, 53.2, 48.3, 55.2, 58.5, 49.0, 46.7, 52.5
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
h = np.array([147.9, 163.5, 159.8, 155.1, 163.3, 158.7, 172.0, 161.2, 153.9, 161.6])
w = np.array([41.7, 60.2, 47.0, 53.2, 48.3, 55.2, 58.5, 49.0, 46.7, 52.5])
```

```python
X = pd.DataFrame(h, columns=["身高"])
y = pd.DataFrame(w, columns=["體重"])
```

```python
lm = LinearRegression()  # 建立線性回歸模型的框架
lm.fit(X, y)
print("迴歸係數: ", lm.coef_)
print("截距: ", lm.intercept_)
```

```python
# 預測氣溫 11, 30 度營業營應該是多少?
new_h = pd.DataFrame(np.array([150, 160, 170]), columns=["身高"])
pred_w = lm.predict(new_h)
print("AI的答案:", pred_w)
```

```python
# 繪圖表示
plt.scatter(h, w)
re_w = lm.predict(X)
plt.plot(h, re_w, color="blue")
plt.plot(new_h, pred_w, color="red", marker="o", markersize=10)
plt.title("Proper Weight for School Student")
```
